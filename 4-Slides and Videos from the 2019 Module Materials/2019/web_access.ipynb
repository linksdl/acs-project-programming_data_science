* Getting the contents of file stored on the web from its url 

```python
import urllib.request

def print_text_from_url( url ):
        url_data = urllib.request.urlopen(url)
        lines = url_data.readlines() # Read all lines into a list
        for line in lines:
                # Decode 'utf-8' string encoding and chop off newline:
                decoded_line = line.decode('utf-8')[0:-1]
                print( decoded_line )
                
def get_url_text( url ):
        url_data = urllib.request.urlopen(url)
        return url_data.read() 

LOREM_IPSUM_URL = "https://teaching.bb-ai.net/PythonCoding/Files_and_Web/lorem_ipsum.txt"

print_text_from_url( LOREM_IPSUM_URL )

# get_url_text( LOREM_IPSUM_URL )

#rawline = urllib.request.urlopen(LOREM_IPSUM_URL).readline()
#print(rawline)

#line = rawline.decode('utf-8')
#print(line)



```

```python
import urllib.request    # used to get data from the web
import csv               # used for reading a csv

QUAKES = "http://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/all_day.csv"

## Give it the URL of a csv file that is available on the internet.
## It will return a datalist of the contents of that file.
def get_datalist_from_csv_url( url ):
    url_response = urllib.request.urlopen(url)
    url_data = url_response.readlines()
    url_data = [line.decode('utf-8') for line in url_data]
    reader = csv.reader(url_data)
    datalist = list(reader)
    return datalist

get_datalist_from_csv_url( QUAKES )
```
